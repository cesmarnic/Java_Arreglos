# Clase 1: Introducción a Arreglos en Java
---

1. Objetivo de la Clase: Comprender qué son los arreglos en Java y cómo se utilizan en la programación.

2. Presentación del Tema:
    * Definición de arreglos y su función en la programación.

        Un arreglo en Java es una estructura de datos que nos permite almacenar un conjunto de valores del mismo tipo bajo un mismo identificador. Los arreglos son útiles para representar datos que tienen una relación lógica entre ellos, como por ejemplo, una lista de nombres, una serie de notas, un conjunto de coordenadas, entre otros. En Java, los arreglos se utilizan para almacenar datos en memoria de manera eficiente y accesible.


    * Características y tipos de arreglos en Java.
    
        En Java, los arreglos tienen las siguientes características:
        * Todos los elementos de un arreglo son del mismo tipo de datos.
        * Los elementos del arreglo están almacenados en posiciones consecutivas de memoria.
        * Los elementos del arreglo se pueden acceder mediante su posición en el arreglo.

        En Java, existen dos tipos de arreglos: unidimensionales y multidimensionales. Los arreglos unidimensionales son aquellos que contienen una sola fila de elementos, mientras que los arreglos multidimensionales son aquellos que contienen varias filas y columnas de elementos. Los arreglos multidimensionales se utilizan para representar datos que tienen una estructura más compleja, como por ejemplo, una matriz de números.

    * Ejemplos de cómo se utilizan los arreglos en la vida cotidiana y en la programación.

        Ejemplos de cómo se utilizan los arreglos en la vida cotidiana y en la programación:
        Los arreglos son muy útiles en la vida cotidiana y en la programación. Por ejemplo, en la vida cotidiana, un arreglo se puede utilizar para almacenar una lista de nombres de personas:
        ```java
        String[] nombres = {"Juan", "Pedro", "Luis", "Ana", "María"};
        ```
        una lista de compras en el supermercado:
        ```java
        String[] listaCompras = {"Leche", "Huevos", "Pan", "Queso", "Jugo"};
        ```
        entre otros. En la programación, los arreglos se utilizan para almacenar datos de entrada de un programa, datos de salida de un programa, matrices para representar imágenes, entre otros.

3. Declaración de arreglos:
    * Sintaxis para declarar arreglos en Java.

        La sintaxis básica para declarar un arreglo en Java es la siguiente:
        ```java	
        tipo[] nombreArreglo;
        ```
        Donde ```tipo``` representa el tipo de datos que se almacenará en el arreglo y ```nombreArreglo``` es el nombre que se le dará al arreglo. Por ejemplo, para declarar un arreglo de enteros en Java, se utilizaría la siguiente sintaxis:
        ```java
        int[] numeros;
        ```

    * Cómo inicializar arreglos con valores predefinidos.
    
        Una vez que se ha declarado un arreglo, se puede inicializar con valores predefinidos utilizando una sintaxis específica. Hay dos formas de inicializar un arreglo con valores predefinidos en Java:

        * Utilizando llaves (```{}```):
            ```java
            int[] numeros = {1, 2, 3, 4, 5};
            ```
            En este ejemplo, se ha creado un arreglo llamado numeros de tipo ```int``` y se ha inicializado con los valores ```1```, ```2```, ```3```, ```4``` y ```5```.
        * Utilizando la clave ```new```:
            ```java
            int[] numeros = new int[5];
            ```
            En este ejemplo, se ha creado un arreglo llamado numeros de tipo ```int``` y se ha inicializado con 5 elementos. Los elementos del arreglo se inicializan con el valor ```0```.
            
    * Ejemplos prácticos de la declaración de arreglos en Java.
        ```java
        // Declaración de un arreglo de enteros sin inicializar
        int[] numeros;

        // Declaración de un arreglo de cadenas con inicialización
        String[] nombres = {"Juan", "María", "Pedro"};

        // Declaración de un arreglo de caracteres sin inicializar
        char[] letras;

        // Declaración de un arreglo de flotantes con inicialización
        float[] precios = {1.99f, 2.99f, 3.99f};

        // Declaración de un arreglo de booleanos sin inicializar
        boolean[] banderas;

        // Declaración de un arreglo de enteros con inicialización
        int[] edades = new int[5];
        ```
